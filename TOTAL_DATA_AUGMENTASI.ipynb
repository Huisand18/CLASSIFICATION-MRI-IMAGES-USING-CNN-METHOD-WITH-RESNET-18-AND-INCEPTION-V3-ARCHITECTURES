# **TOTAL DATA AUGMENTASI**

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import shutil

# Fungsi untuk membuat direktori baru
def create_dir(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

# Path untuk menyimpan gambar augmentasi
augmented_train_dir = "/content/drive/MyDrive/SKRIPSI/dataset_augmented/train"
augmented_val_dir = "/content/drive/MyDrive/SKRIPSI/dataset_augmented/val"
augmented_test_dir = "/content/drive/MyDrive/SKRIPSI/dataset_augmented/test"

# Membuat direktori untuk menyimpan gambar augmentasi
create_dir(augmented_train_dir)
create_dir(augmented_val_dir)
create_dir(augmented_test_dir)

# ImageDataGenerator untuk augmentasi
augmentor = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    rotation_range=45,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)

# Fungsi untuk menyimpan gambar augmentasi
def augment_and_save_images(generator, source_dir, dest_dir, batch_size=64, img_height=299, img_width=299, num_augmented_images_per_class=5):
    for class_name in os.listdir(source_dir):
        class_path = os.path.join(source_dir, class_name)
        if os.path.isdir(class_path):
            dest_class_path = os.path.join(dest_dir, class_name)
            create_dir(dest_class_path)
            images = os.listdir(class_path)
            for img_name in images:
                img_path = os.path.join(class_path, img_name)
                img = cv2.imread(img_path)
                if img is not None:
                    img = cv2.resize(img, (img_width, img_height))
                    img = img / 255.0  # Normalisasi
                    img = np.expand_dims(img, axis=0)
                    aug_iter = generator.flow(img, batch_size=1)
                    for i in range(num_augmented_images_per_class):
                        aug_img = next(aug_iter)[0]
                        aug_img = (aug_img * 255).astype(np.uint8)
                        aug_img_path = os.path.join(dest_class_path, f"aug_{i}_{img_name}")
                        cv2.imwrite(aug_img_path, aug_img)

# Mengaugmentasi dan menyimpan gambar untuk train, val, dan test
augment_and_save_images(augmentor, train_dir, augmented_train_dir)
augment_and_save_images(augmentor, val_dir, augmented_val_dir)
augment_and_save_images(augmentor, test_dir, augmented_test_dir)

# Fungsi untuk menghitung jumlah gambar dalam direktori
def count_files_in_directory(directory):
    total_count = 0
    class_counts = {}
    for root, dirs, files in os.walk(directory):
        class_name = os.path.basename(root)
        class_counts[class_name] = len(files)
        total_count += len(files)
    return total_count, class_counts

# Menghitung jumlah gambar setelah augmentasi
aug_train_total, aug_train_class_counts = count_files_in_directory(augmented_train_dir)
aug_val_total, aug_val_class_counts = count_files_in_directory(augmented_val_dir)
aug_test_total, aug_test_class_counts = count_files_in_directory(augmented_test_dir)

print("Jumlah data/gambar setelah augmentasi di setiap path:")
print(f"Train: {aug_train_total}")
print(f"Test: {aug_test_total}")
print(f"Validation: {aug_val_total}")

print("\nJumlah data/gambar setelah augmentasi dalam setiap kelas:")
print("Train:")
for class_name, count in aug_train_class_counts.items():
    print(f"{class_name}: {count}")

print("\nTest:")
for class_name, count in aug_test_class_counts.items():
    print(f"{class_name}: {count}")

print("\nValidation:")
for class_name, count in aug_val_class_counts.items():
    print(f"{class_name}: {count}")
```

# **AUGMENTASI DATA**

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    rotation_range=45,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.4,
    zoom_range=0.2,
    vertical_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(
    rescale = 1./255,
)
```

```python
img_height = 299
img_width = 299
batch_size= 64
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    shuffle=True,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    shuffle=True,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode = 'categorical',
    shuffle = False,
)
```

# **MODELING**

## **INCEPTION V3**

```python
pre_model = InceptionV3(
        input_shape=(299,299,3),
        weights='imagenet',
        include_top=False,
)
for layers in pre_model.layers:
    layers.trainable=False

last_layer = pre_model.get_layer('mixed7')
last_output = last_layer.output
```

```python
x = tf.keras.layers.Flatten()(last_output)
x = tf.keras.layers.Dense(512,activation='relu')(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dense(256,activation='relu')(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dense(128,activation='relu')(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dense(64,activation='relu')(x)
x = tf.keras.layers.BatchNormalization()(x)
output = tf.keras.layers.Dense(4,activation='softmax')(x)

model = Model(pre_model.input,output)

model.compile(optimizer=Adam(learning_rate=0.0001),
                loss = 'categorical_crossentropy',
                metrics=['accuracy']
)
```
